```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Aufgabe 1 (a)

L =np.array([[1,0,0],[-2,1,0],[4,5,1]])
R =np.array([[2,-1,6],[0,3,9],[0,0,-2]])
b= np.array([18,-3,231])
    
def vorrueck(L,R,b):

#Vorwärtseinsetzen
    
    y1 = 0
    y2 = 0
    y3 = 0
    y1 = b[0]/L[0,0]
    y2 = (b[1]+2*y1)/L[1,1]
    y3  = (b[2] -4*y1-5*y2)/L[2,2]
    y = np.array([y1 ,y2,y3])

#Rückwärtseinsetzen
    
    x1 = 0
    x2 = 0 
    x3 = 0
    x3 = y[2]/R[2,2]
    x2 = (y[1]-R[1,2]*x3)/R[1,1]
    x1 = (y[0]-R[0,1]*x2-R[0,2]*x3)/R[0,0]
    x = np.array([x1,x2,x3])
    print('x = ',x.T)
    return 
    
vorrueck(L,R,b)
```

```python
# Aufagebe 1 (b)

A = np.matrix([[2,4,5,-3],[10,19,27,-14],[6,16,11,-11],[-4,-11,4,19]])
def lu(A):
  
    n= A.shape[0]
    R = A
    L = np.array([[1 if i == j else 0 for j in range(n)] for i in range(n)])
    for j in range(0,n-1):
        for i in range(j+1,n):
            m = A[i,j]/A[j,j]
            R[i,j+1:n]= R[i, j+1:n] - m * R[j,j+1:n]
            L[i,j] = m
            R[i,j] = 0
    return L ,R 




L ,R = lu(A)
print('L=\n',L)
print('R=\n',R)

    

    



















```

```python
#Aufgabe 1 (c) I


x =np.linspace(0,1,1000)
y1 = np.sin(5*x)
y2=np.cos(5*x)

plt.figure(figsize=(15,4))
plt.xlim(0,1)
plt.ylim(-2,2)
plt.xticks(np.arange(0,1.1,0.1))
plt.yticks(np.arange(-2,3,1))
plt.tick_params(axis='both',direction ='in',top=True,right= True)           
plt.grid(True,which='both',linestyle= '--',linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=sin(5x) und y=cos(5x)')
plt.plot(x,y1,'k',linewidth=0.7)
plt.plot(x,y2,'k',linewidth=0.7)
plt.show
```
